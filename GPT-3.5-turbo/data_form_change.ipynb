# 1. 경로 지정

```python
import os
import json

input_folder= "/content/drive/MyDrive/DILAB/YB/GPT_FineTuning/news_data" # 원본 JSONL 파일이 들어있는 폴더
output_file = "/content/drive/MyDrive/DILAB/YB/GPT_FineTuning/GPT-3.5-turbo/finetune_data.jsonl" # 변환 결과를 저장할 최종 파일 경로
```

# 2. 변환 함수 정의

```python
import json

def convert_record_with_fixed_question(record): # 기존 뉴스데이터를 OpenAI 파인튜닝용 대화 형식으로 변환
    """
    기존 record: {"title": "...", "body": "..."}
    변환 후: {"messages": [{"role":"user",...}, {"role":"assistant",...}]}
    """
    return { # OpenAI Chat 파인튜닝 포맷으로 반환
        "messages": [
            {"role": "user", "content": "최근 뉴스를 알려줘."},
            {"role": "assistant", "content": record.get("body", "").strip()} # body키가 없을 경우 빈 문자열로 대체, 양쪽 공백 제거(불필요한 개행/공백 정리)
        ]
    }
```

# 3. 데이터폼 변환

```python
import os

with open(output_file, "w", encoding="utf-8") as fout: # 출력 파일을 쓰기 모드로 open
    for filename in os.listdir(input_folder): # 입력 폴더 내 모든 파일을 순회
        if filename.endswith(".jsonl"): # 확장자가 .jsonl인 파일만 처리
            file_path = os.path.join(input_folder, filename)
            with open(file_path, "r", encoding="utf-8") as fin: # 각 파일을 한 줄씩 읽어서 반환
                for line in fin:
                    if not line.strip(): # 공백/개행만 있는 줄은 처리하지 않음
                        continue
                    record = json.loads(line)
                    new_record = convert_record_with_fixed_question(record) # 변환 수행
                    fout.write(json.dumps(new_record, ensure_ascii=False) + "\n") # 변환 결과를 기록

print(f"변환 완료: {output_file}")
```
